#### Exploratory Data Analysis
There are 618 rows and 12 columns in this dataset. The columns represent the physical properties and chemical composition of soil. 
1. Area 
2. Data Collection Year
3. Soil group
4. Land class
5. Knit (surface)
6. pH
7. SOC (%)
8. Nitrogen N (%)
9. Potassium K (meq/100)
10. Phosphorus P (ug/g)
11. Sulfur S (ug/g)
12. Boron B (ug/g)
13. Zinc Zn (ug/g)


```python
for i in range(5,len(df.columns)):
    df[df.columns[i]] = pd.to_numeric(df[df.columns[i]],errors='coerce')
```

```python
# Analysing the categorical variables
for col in df.columns:
    if df[col].dtype in [object]:
        print('Attribute :',col)
        print(df[col].describe())
        print(df[col].unique())
        print('-----------------------------------')
```

```python
# Analysing the numerical columns
for col in df.columns:
    if df[col].dtype in [float]:
        print('Attribute :',col)
        print(df[col].describe())
        print(df[col].info())
        print('-----------------------------------------')
```

```python
# Removing outliers in the data
df = df[(df['Nitrogen N (%)']<0.8)&(df['Potassium K (meq/100)']<10)&(df['SOC (%)']<5)&(df['Boron B (ug/g)']<5)]
```

```python
# checking the linear correlation between the variables
df_corr = df.corr(numeric_only=True)
sns.heatmap(df_corr,cmap="PiYG",vmin=-1, vmax=1, center=0)
```
